```python
# creating dataframe using dictionary
data = {
    "name": ["Arun", "John", "Meera"],
    "age": [21, 22, 23],
    "marks": [85, 90, 95]
}

df = pd.DataFrame(data)
print(df.head())
```

```python
# filtering the people who has >= 90 marks
df.loc[df["marks"] >= 90]
```

```python
# same as above
df[df["marks"] >= 90]
```

```python
# filtering the people who's age is less than 22
df[df["age"] < 22]
```

```python
# creating a new column, storing result of condition >= 90 marks
df["res"] = df["marks"] >= 90
```

```python
df.head()
```

```python
# creating a new column to make age group, by storing result of apply function to implement a function
# a lambda function is similar to an udf, which can take multiple arguments but single statement.
df["age_group"] = df["age"].\
    apply(lambda x: "Adult" if x > 19 else "Teen")
```

```python
# creating 2 dataframes
students = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Arun", "John", "Meera"]
})

marks = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "marks": [85, 90, 95, 100]
})

#performing merging with inner join based on id column

merged = pd.merge(students, marks, on = "id", how = "inner")
print(merged.head())
```

```python
import pandas as pd

# creating a dataframe for sales data

df = pd.DataFrame({
    "city": ["Tashkent", "Tashkent", "Bukhara", "Bukhara"],
    "sales": [100, 150, 120, 130]
})

print(df.head())
```

```python
# groupby city to find total sales of each city
df.groupby("city")["sales"].sum()
```

```python
df.groupby("city")["sales"].mean()
```

```python
# groupby city to find sum, average, max, and min of sales
df.groupby("city")["sales"].agg(["sum", "mean", "max", "min"])
```

```python
# finding 5 point summary, count, mean, and standard deviation of the dataframe
df.describe()
```
